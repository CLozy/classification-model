```python
features = diabetes_data.drop('test', axis=1)
features.head()
```

```python
from sklearn import preprocessing
standard_scaler = preprocessing.StandardScaler()
```

```python
features_scaled = standard_scaler.fit_transform(features)

features_scaled.shape
```

```python
features_scaled_df = pd.DataFrame(features_scaled, columns=features.columns)

features_scaled_df.head()
```

```python
features_scaled_df.describe()
```

```python
diabetes_data = pd.concat([features_scaled_df, diabetes_data['test']], axis=1).reset_index(drop=True)

    
diabetes_data.head()   
```

```python
diabetes_data.to_csv(r'C:\Users\Shaleen\Desktop\jpnb\DESIGNING\2.0\PimaIndians_processed.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split

X = diabetes_data.drop('test', axis=1)
Y = diabetes_data['test']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
 x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

#fits an s-curve on underlying data
#classifies categories dependin on threshold
#l2 regularization technique to prevent module from overfitting on training data
#solver='liblinear is optimization algorithm that LogisticRegression will use underthehood'
classifier = LogisticRegression(penalty='l2', C=1.0, solver='liblinear')

classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)

y_pred
```

```python
#dataframe that compares actual labels from predicted labels
pred_results = pd.DataFrame({'y_test': y_test,
                            'y_pred': y_pred})

pred_results.head(10)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
#how many of the model's classifications were actually correct
model_accuracy = accuracy_score(y_test, y_pred)
#how many of the model's positive classifications of diabetes were correct
model_precision = precision_score(y_test, y_pred)
#how many of the individuals with diabetes were correctly classified by model 
model_recall = recall_score(y_test, y_pred)

print('Accuracy of the model is {}%' .format(model_accuracy * 100))
print('Precision of the model is {}%' .format(model_precision * 100))
print('Recall of the model is {}%' .format(model_recall * 100))
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(max_depth=4)

classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)

y_pred
```

```python
#dataframe that compares actual labels from predicted labels
pred_results = pd.DataFrame({'y_test': y_test,
                            'y_pred': y_pred})

pred_results.head(10)
```

```python
#how many of the model's classifications were actually correct
model_accuracy = accuracy_score(y_test, y_pred)
#how many of the model's positive classifications of diabetes were correct
model_precision = precision_score(y_test, y_pred)
#how many of the individuals with diabetes were correctly classified by model 
model_recall = recall_score(y_test, y_pred)

print('Accuracy of the model is {}%' .format(model_accuracy * 100))
print('Precision of the model is {}%' .format(model_precision * 100))
print('Recall of the model is {}%' .format(model_recall * 100))
```

```python
#view actual label vs predicted values
diabetes_crosstab = pd.crosstab(pred_results.y_pred, pred_results.y_test)

diabetes_crosstab 
```

```python
#calculate manually accuracy, precision and recall scores
TP = diabetes_crosstab[1][1]
TN = diabetes_crosstab[0][0]
FP = diabetes_crosstab[0][1]
FN = diabetes_crosstab[1][0]
```

```python
accuracy_score_verified = (TP + TN) / (TP + FP + TN + FN)

accuracy_score_verified
```

```python
precision_score_survived = TP / (TP + FP)

precision_score_survived
```

```python
recall_score_survived = TP / (TP + FN)
recall_score_survived
```
